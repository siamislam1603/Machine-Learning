### Put the API key

```python
import requests,json,time
import pandas as pd
api_key=""
```

```python
restaurants_dic={'Name':[],'Latitude':[],'Longitude':[],'Rating':[],'User Ratings Total':[],'Price':[]}
coordinates = ['20.566667,88.016667']
for coordinate in coordinates:
    url = f'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={coordinate}&radius=25000\
    &keyword=restaurant&key={api_key}'
    while True:
        response = requests.get(url)
        data = json.loads(response.text)
        for info in data['results']:
            restaurants_dic['Name'].append(info['name'])
            restaurants_dic['Latitude'].append(info['geometry']['location']['lat'])
            restaurants_dic['Longitude'].append(info['geometry']['location']['lng'])
            restaurants_dic['User Ratings Total'].append(info['user_ratings_total'])
            restaurants_dic['Rating'].append(info['rating'])
            if 'price_level' in info.keys():
                restaurants_dic['Price'].append('$'*int(info['price_level']))
            else:
                restaurants_dic['Price'].append(None)
        time.sleep(5)
        if 'next_page_token' not in data.keys():
            break
        else:
            url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?key={api_key}\
            &pagetoken={data['next_page_token']}"
    try:
        if(restaurants_dic['Latitude'][-1]<26.633333 and restaurants_dic['Longitude'][-1]<92.683333):
            coordinates.append(str(restaurants_dic['Latitude'][-1])+','+str(restaurants_dic['Longitude'][-1]))
        else:
            break
    except:
        cord=coordinates[-1].split(',')
        coordinates.append(str(float(cord[0])+1)+','+str(float(cord[1])+1))
```

```python
df=pd.DataFrame(restaurants_dic)
df=df.drop_duplicates().reset_index(drop=True)
df
```

```python
df.to_csv('Restaurants_info.csv',index=False)
```
