```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
x=DataFrame(data,columns=['production_budget_usd'])
y=DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
plt.title('Film budget vs Global revenue')
plt.xlabel('Production budget $')
plt.ylabel('Worldwide gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression=LinearRegression()
regression.fit(x,y)
```

Slope coefficient:

```python
regression.coef_ # theta_1 or slope of linear line
```

```python
regression.intercept_ #theta_0 or constant of linear line
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3) # scatter all points in the graph & alpha is fr transparency
plt.plot(x,regression.predict(x),color='red',linewidth=4) # plot a line with predicted value for x for all x value

plt.title('Film budget vs Global revenue')
plt.xlabel('Production budget $')
plt.ylabel('Worldwide gross $')
plt.ylim(0,3000000000) #limitation of y axis
plt.xlim(0,450000000) #limitation of x axis
plt.show()
```

```python
regression.score(x,y)
```
