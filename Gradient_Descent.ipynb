```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example 1 - A simple cost function
## $f(x)=x^2+x+1$

```python
def f(x):
    return x**2+x+1
```

```python
x_1=np.linspace(start=-3,stop=3,num=100)
x_1
```

```python
#plot function
plt.plot(x_1,f(x_1))
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.show()
```

## Slope & Derivatives

```python
def df(x):
    return 2*x+1
```

```python
# plot function and derivative side by side

plt.figure(figsize=[15,5])

# 1 chart:cost function:
plt.subplot(1,2,1)
plt.plot(x_1,f(x_1),color='blue',linewidth=3)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.title('Cost function')

# 2 Chart:Dervative
plt.subplot(1,2,2)
plt.plot(x_1,df(x_1),color='skyblue',linewidth=5)
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.grid()
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.title('slope of the cost function')

plt.show()
```

```python
# Gradient Descent

new_x=3
previous_x=0
step_multiplier=0.1
precision=0.00001
x_list=[new_x]
slope_list=[df(new_x)]

for n in range(500):
    previous_x=new_x
    gradient=df(previous_x)
    new_x=previous_x-(step_multiplier*gradient)
    x_list.append(new_x)
    slope_list.append(df(new_x))
    if(abs(new_x-previous_x)<precision):
        print('Loop ran ',n,' times')
        break
print('Local minimum occurs at:',new_x)
print('Slope or df(x) value at this point is:',df(new_x))
print('f(x) or cost value at this point:',f(new_x))
```

```python
# superimpose the plot function and derivative side by side

plt.figure(figsize=[20,5])

# 1 chart:cost function:
plt.subplot(1,3,1)
plt.plot(x_1,f(x_1),color='blue',linewidth=3,alpha=0.8)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.title('Cost function')
values=np.array(x_list)
plt.scatter(x_list,f(values),color='red',s=100,alpha=0.5)

# 2 Chart:Dervative
plt.subplot(1,3,2)
plt.plot(x_1,df(x_1),color='skyblue',linewidth=5,alpha=0.6)
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.grid()
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.title('slope of the cost function')
plt.scatter(x_list,slope_list,color='red',s=100,alpha=0.5)

# 2 Chart:Gradient Descent(close up)
plt.subplot(1,3,3)
plt.plot(x_1,df(x_1),color='skyblue',linewidth=5,alpha=0.6)
plt.xlim(-0.6,2.5)
plt.ylim(-0.5,5.8)
plt.grid()
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.title('Gradient Descent(close up)')
plt.scatter(x_list,slope_list,color='red',s=100,alpha=0.5)

plt.show()
```
