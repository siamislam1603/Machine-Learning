```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

#### Subplots in the same canvas- subplot(rows,cols,index)

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

#### Object oriented base method for plotting-add_axes([left,bottom,width,height])

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y,'r')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Larger Plot')

axes2=fig.add_axes([0.2,0.4,0.3,0.4])
axes2.plot(y,x,'g')
axes2.set_ylabel('X Label')
axes2.set_xlabel('Y Label')
axes2.set_title('Larger Plot')
```

### Subplots

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout()
```

#### Figure Size and DPI

```python
fig=plt.figure(figsize=(8,2)) # figsize=(width,height) in inches
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='X^2')
axes.plot(x,x**3,label='X^3')
plt.legend(loc=0)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```
