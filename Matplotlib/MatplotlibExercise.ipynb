```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y,'r')
ax2.plot(x,y,'r')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig=plt.figure()

ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,y,'b',label='y')
ax1.plot(x,z,'g',label='z')
ax2.plot(x,y,'b')

ax1.set_xlim([0,100])
ax1.set_ylim([0,10000])

ax2.set_xlim([20,22])
ax2.set_ylim([30,50])

ax1.set_xlabel('x')
ax1.set_ylabel('z')

ax2.set_xlabel('x')
ax2.set_ylabel('y')

ax1.legend()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,color='blue',lw=3,ls='--')
axes[1].plot(x,z,color='red',lw=5,ls='-')

axes[0].set_xlim([0,100])
axes[0].set_ylim([0,200])

axes[1].set_xlim([0,100])
axes[1].set_ylim([0,10000])

plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,3))

axes[0].plot(x,y,color='blue',lw=5,ls='-')
axes[1].plot(x,z,color='red',lw=3,ls='--')

axes[0].set_xlim([0,100])
axes[0].set_ylim([0,200])

axes[1].set_xlim([0,100])
axes[1].set_ylim([0,10000])

plt.tight_layout()
```
